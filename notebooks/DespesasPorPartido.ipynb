```python
import pandas as pd
import glob

df = pd.DataFrame({})
files = glob.glob('../build/*.csv')
for file in files:
    df = df.append(pd.read_csv(file, sep=';'))
```

```python
groupedExpenses = df[['sgPartido', 'vlrLiquido']].set_index('sgPartido').groupby('sgPartido')
```

```python
expensesByParty = groupedExpenses.sum().sort_values('vlrLiquido', ascending=False).astype('int64')

expensesByParty.assign(totalExecutado=groupedExpenses.count())
```
